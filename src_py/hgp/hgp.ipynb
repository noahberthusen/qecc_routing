```python
import numpy as np
import matplotlib.pyplot as plt
import galois
from numpy.linalg import matrix_power, matrix_rank
import itertools
```

```python
def shift_right(arr, i):
    return arr[-i:] + arr[:-i]

def generate_H(p, n, k):
    first_row = [0 for _ in range(n)]

    for i, term in enumerate(p):
        if term: first_row[i] = 1

    H = [first_row]
    for i in range(1,n-k):
        H.append(shift_right(first_row, i))

    return np.array(H)

def find_distance(H):
    n = H.shape[1]
    d = n

    for j in range(1, 2**n):
        num = bin(j)[2:].zfill(n)
        w = np.array([int(b) for b in num])
        synd = (H@w) % 2

        if (not np.any(synd)):
            if (np.count_nonzero(w) < d):
                d = np.count_nonzero(w)
    return d

GF = galois.GF(2)
def get_dimension(H):
    n = H.shape[1]
    rank = matrix_rank(GF(H))
    return n-rank


p = [1,0,1,1]
n = 13

params = []
for i in range(len(p), n):
    tmp = []
    for j in range(i):
        H = generate_H(p, i, j)

        # d = 0
        # for comb in itertools.combinations(np.arange(i), j+1):
        #     tmp_H = H[list(comb)]
        #     tmp_d = find_distance(tmp_H)
        #     if tmp_d > d:
        #         d = tmp_d
        # tmp.append(d)
        tmp.append((get_dimension(H), find_distance(H), get_dimension(H.T), find_distance(H.T)))

    print(i, tmp)
    # params.append(tmp)
```

```python
def generate_binary_strings(n):
    seen = set()
    for i in range(1,2**n):
        binary = bin(i)[2:][::-1].strip('0')
        reverse = binary[::-1]
        if reverse not in seen and binary not in seen:
            seen.add(binary)
            yield binary
```

```python
n = 15
nkd = np.zeros(shape=(n,n),dtype=int)
codes = np.zeros(shape=(n,n),dtype=object)

for p in generate_binary_strings(10):
    if p.count('1') > 4: continue
    p_int = int(p,2)
    w = [int(b) for b in p]
    print(w)

    for i in range(len(w), n):
        for j in range(i):
            H = generate_H(w, i, j)

            k = get_dimension(H)
            d = find_distance(H)

            if d > nkd[i][k]:
                nkd[i][k] = d
                codes[i][k] = (p_int,j)
```
